### Our objective

$$
\newcommand{\BS}{\text{BS}}
$$

- Our objective is to approximate the shape of the implied volatility surface for any given stochastic volatility model, the Heston model in particular.


- In Lecture 2, we showed how to compute local volatilities from implied volatilities:

    <a name="eq1:localFromImplied"></a>(1)
$$
\begin{eqnarray*}
v_\ell(k,T) = \frac{\frac{\partial w}{\partial T}}
{\left(1 -\frac 12\frac{k}{w}\,\frac{\partial w}{\partial k}\right)^2
 - \frac14\,\left( \frac{1}{4} + \frac{1}{w} \right)\,
 \left( \frac{\partial w}{\partial k} \right)^2  +
\frac{1}{2}\,\frac{\partial^2 w}{\partial k^2 }}
\end{eqnarray*}
$$

- Recall that given a model, we may approximate the conditional expectation of $v_t$ to get $v_\ell$ using
    
    <a name="eq2:localvolisexpectedv"></a>(2)
    $$v_\ell(k,T)=\mathbb{E}[v_t|x_T=k].$$  


- An obvious approach would then be to compute local volatility in the stochastic volatility model, then invert [(1)](#eq1:localFromImplied) to get implied volatility.

    - <span>We showed in Lecture 2 how to invert [(1)](#eq1:localFromImplied) for small times to
expiration.</span>

    - <span>However, nobody has yet figured out how to do this in general!</span>

- <span>Instead, by exploiting the insight (originally due to Bruno
Dupire) that implied volatility is related to the breakeven level of a
delta-hedging strategy, we derive a general path-integral representation
of Black-Scholes implied variance.</span>

### Delta hedging under stochastic volatility

Assume the stock price $S_t$ satisfies the SDE

$$\frac{{dS_t }}{{S_t }} = \mu _t dt + \sigma _{t } \,dZ_t$$ 

where the
volatility $\sigma _{t }$ may be random.
Then path-by-path, for any suitably smooth function $f\left( {S_t ,t}\right)$,


\begin{eqnarray*}
&&f\left( {S_T ,T} \right) - f\left( {S_0 ,0} \right) \\&=& \int_0^T
{df}\nonumber \\ &=& \int_0^T {\left\{ {\frac{{\partial
f}}{{\partial S_t }}\,dS_t  + \frac{{\partial f}}{{\partial
t}}\,dt + \frac{{\sigma _{t} ^2 }}{2}\, S_t ^2\,\frac{{\partial ^2
f}}{{\partial S_t ^2 }}\,dt} \right\}} \end{eqnarray*}

In particular (with $f(\cdot)=C_{BS}(\cdot)$), the value of a call
option is given by:

$$
\begin{eqnarray*}
&& C\left( {S_0,K,T} \right)\\ &=& \mathbb{ E}\left[ {\left( {S_T  -
K} \right)^+  } \right] \nonumber\\ &=&\mathbb{ E}\left[ {C_{BS} \left( {S_T, K,\bar\sigma(T),0}
\right)} \right] \nonumber \\ &=&
C_{BS} \left( {S_0,K ,\bar\sigma(0),T} \right)\nonumber\\
&& +
\mathbb{ E}\left[ { {\int_0^T {\left\{ {\frac{{\partial
C_{BS} }}{{\partial S_t }}dS_t  + \frac{{\partial C_{BS}
}}{{\partial t}}dt + \frac{1}{2}\sigma _{t} ^2 S_t ^2
\frac{{\partial ^2 C_{BS} }}{{\partial S_t ^2
}}dt} \right\}} }   } \right]
\end{eqnarray*}
$$

where $\bar \sigma(t),\,t \in [0,T]$ is some bounded positive function of $t$.

### The breakeven computation

Now let 

$$
\bar \sigma(t)^2 = \frac{1}{T-t}\, \int_t^T\,v(s)\,ds
$$

for some deterministic (bounded and non-negative) forward variance function $v(\cdot)$.  

Then 
$C_{BS} \left(S_t,K,\bar \sigma(t),T-t \right)$  satisfies the
Black-Scholes equation (assuming zero interest rates and dividends):

$$
\frac{{\partial C_{BS} }}{{\partial t}} =  -
\frac{1}{2}\,v(t)\,S_t ^2 \,\frac{{\partial ^2
C_{BS} }}{{\partial S_t ^2 }}.
$$ 

Substituting for
the time derivative $\frac{{\partial C_{BS} }}{{\partial t}}$, and using that $\mathbb{E}[dS_t]=0$, we
obtain:

$$
\begin{eqnarray*}
&&C\left( {S_0,K,T} \right)\\  
&=& C_{\BS} \left( {S_0,K ,\bar\sigma(0),T} \right) \nonumber\\&&+ \mathbb{
E}\left[ { {\int_0^T {\frac{1}{2}\left\{ \sigma _{t} ^2  -
v(t) \right\}S_t ^2 \frac{{\partial ^2
C_{BS} }}{{\partial S_t ^2 }}dt} }  }
\right]. \end{eqnarray*}
$$

In particular, if we choose $\bar \sigma(0) = \sigma_{\BS}(k,T)$, we have 
$
C\left( {S_0,K,T} \right) = C_{\BS} \left( {S_0,K ,\bar\sigma(0),T} \right) 
$
and

<a name="eq:breakeven"></a>(3)
$$
\mathbb{
E}\left[ { {\int_0^T {\frac{1}{2}\left\{ \sigma _{t} ^2  -
v(t) \right\}S_t ^2 \frac{{\partial ^2
C_{BS} }}{{\partial S_t ^2 }}dt} }  }
\right]=0.
$$

### Representation of implied volatility

We then have the following lemma (from [Guyon and Henry-Labordère]<sup id="cite_ref-GHL" class="reference"><a href="#cite_note-GHL"><span>[</span>5<span>]</span></a></sup>  ):
<blockquote><div style="background-color:#add8e6; color:#FFFFFF; font-style: normal;  " ><h4>
Lemma 1 (Guyon and Henry-Labordère)</h4>
</div>
<div style="background-color:#E8E8E8; color:#000000; font-style: normal; ">
<br>
<a name="eq:lemma1"></a>(4)
$$\sigma_{BS}^2(K,T)=\frac{1}{T}\,\int_0^T\,v(t)\,dt$$ if and only if
$$\int_0^T 
{\mathbb{E} }\left[ \frac12\left\{ \sigma _t^2  -
v(t) \right\}S_t ^2\, \frac{{\partial}^2
C^v_{BS}}{{\partial}S_t ^2}\right]\,dt =0.$$ 


where $C^v_{BS}$ represents
the Black-Scholes formula computed with $v(t)$ as the forward variance
function. 


<br>
<br>

</div>

</blockquote>

<br>



### $\sigma_{BS}$ as the breakeven implied volatility

- In words, 
$$\int_0^T 
{\mathbb{E} }\left[ \frac12\left\{ \sigma _t^2  -
v(t) \right\}S_t ^2\, \frac{{\partial}^2
C^v_{BS}}{{\partial}S_t ^2}\right]\,dt $$ 
gives the
expected realized profit on the purchase of a European call or put option at the market price, delta-hedged using the deterministic
forward variance function $v(t)$ when the actual realized volatility
is $\sigma_t$. 

### Black-Scholes gamma

For fixed $K$ and $T$, define the Black-Scholes gamma

$$\Gamma_{\BS} (S_t,t): = \frac{{\partial ^2
}}{{\partial S_t ^2 }}\,C_{\BS}(S_t,K,\bar \sigma(t),T-t).$$ 

Then, in terms of $\Gamma_{\BS}$, the breakeven condition becomes

$$
\mathbb{
E}\left[ { {\int_0^T {\frac{1}{2}\left\{ \sigma _{t} ^2  -
v(t) \right\}S_t ^2 \,\Gamma_{\BS} (S_t,t )\,dt} }  }
\right]=0.
$$

### Black-Scholes forward implied variance

Now choose $v(t)$ to be the *Black-Scholes forward implied variance* function

<a name="eq:varformula"></a>(5)
$$v(t) = \frac{{\mathbb{ E}\left[ {\sigma _{t} ^2\, S_t ^2
\,\Gamma _{BS} (S_t,t )} \right]}}{{\mathbb{ E}\left[ {S_t ^2\, \Gamma _{BS}
(S_t,t )} \right]}}.
$$

Then

$${\mathbb{ E}\left[ {S_t ^2\, \Gamma _{BS} (S_t,t
)} \right]}\,v\left( t
\right) = {{\mathbb{ E}\left[ {\sigma _{t} ^2\, S_t ^2 \,\Gamma
_{BS} (S_t,t)}
\right]}}$$ 

and the breakeven condition [(4)](#eq:breakeven) is satisfied.


### $\sigma_{BS}^2$ as a path integral

It then follows from Lemma 1 [(5)](#eq:lemma1) that

<blockquote><div style="background-color:#add8e6; color:#FFFFFF; font-style: normal;  " ><h4>
Path integral representation of Black-Scholes implied volatility</h4>
</div>
<div style="background-color:#E8E8E8; color:#000000; font-style: normal; ">
<br>
<a name="eq:explicitimpvol"></a>(6)
$$\sigma_{BS}(K,T)^2 =\frac {1}{T} \int_0^T
{\frac{{\mathbb{ E}\left[ {\sigma _{t} ^2 S_t ^2 \Gamma _{BS} (S_t,t
)} \right]}}{{\mathbb{ E}\left[
{S_t ^2 \Gamma _{BS} (S_t ,t)}
\right]}}dt}$$

<br>
<br>

</div>

</blockquote>

- <span> Equation [(6)](#eq6:explicitimpvol) expresses implied variance as the
time-integral of expected instantaneous variance $\sigma_t^2$ under some
probability measure.</span>

- <span> [(6)](#eq6:explicitimpvol) is <span>*exact*</span> but implicit.</span>

### Interpretation


- To compute the Black-Scholes implied volatility of an option, we
need to average the possible realized volatilities over all possible
scenarios.



- Each such scenario is weighted by the gamma of the option.

    - The profitability of the delta-hedger in any time interval is
directly proportional to the gamma and the difference between “expected
instantaneous variance” (or local variance) and realized instantaneous
variance.

- Only paths that start at $S_0$ and end at the strike price $K$
need be included in the average.

    - At inception of the delta-hedge, the stock price can only be
$S_0$.

    - At expiration, gamma is precisely zero if $S_T \neq K$. Gamma
explodes at $S_T=K$.

### An elegant recasting of [(6)](#eq:explicitimpvol)

Following Roger Lee, we may rewrite [(6)](#eq:explicitimpvol) more elegantly
as

<a name="eq:gdefn"></a>(7)
$$\sigma_{BS}(K,T)^2=\bar\sigma(0)^2=\frac {1}{T}
\int_0^T\,\mathbb{E}^{G_t}[\sigma_t^2]\,dt$$

thus interpreting the definition [(5)](#eq:varformula) of $v(t)$ as the
expectation of $\sigma_t^2$ with respect to the probability measure
$\mathbb{G}_t$ defined, relative to the pricing measure $\mathbb{Q}$, by
the Radon-Nikodym derivative


$$\frac{d\mathbb{G}_t}{d\mathbb{Q}}:=\frac{{ { S_t ^2 \,\Gamma _{BS}
(S_t,\bar \sigma(t) )}}}{{\mathbb{ E}\left[ {S_t ^2\, \Gamma _{BS}
(S_t,\bar \sigma(t) )} \right]}}.$$

<font color="red">

#### <font color="red"><span>Remark</span>

Equations [(5)](#eq:varformula) and [(6)](#eq:explicitimpvol) are implicit
because the gamma $\Gamma
_{BS} (S_t ,t)$ of the option depends on all the forward implied variances
$v\left( t \right)$.

### Special Case: Black-Scholes

Suppose $\sigma _{t} =
\sigma(t)  $, a function of $t$ only. Then

$$v\left( t \right) = \frac{{\mathbb{ E}\left[ {\sigma (t) ^2
S_t ^2 \Gamma _{BS} (S_t )}
\right]}}{{\mathbb{ E}\left[ {S_t ^2 \Gamma _{BS} (S_t )} \right]}} = \sigma (t) ^2.$$

The forward implied variance $v(t)$ and the forward variance
$\sigma(t)^2$ coincide.

As expected,in this case, $v(t)$ has no dependence on the strike $K$ or the
option expiration $T$.

### Understanding the integral representation [(6)](#eq6:explicitimpvol)

In order to get better intuition for the path integral representation [(6)](#eq:explicitimpvol), first
recall how to compute a risk-neutral expectation:

$$\mathbb{ E}^{\mathbb{Q}}\left[ {f\left( {S_t } \right)} \right] = \int
{dS_t \,p\left( {S_t,t ;S_0 } \right)} \,f\left( {S_t } \right).$$ 

We get
the risk-neutral pdf of the stock price at time $t$ by taking the second
derivative of the market price of European options with respect to
strike price.

$$p\left( {S_t,t ;S_0 } \right) = \left. {\frac{{\partial ^2 C\left(
{S_0 ,K,t} \right)}}{{\partial K^2 }}} \right|_{K = S_t }.$$

Then from equation [(7)](#eq:gdefn) we have

<a name="eq8:brownianbridgeS"></a>(8)
\begin{eqnarray*}
v\left( t \right) &=&\mathbb{ E}^{{G_t}}\left[
\sigma_t^2  \right]\nonumber\\
&=& \mathbb{ E}^{\mathbb{Q}}\left[\sigma_t^2 \frac{d\mathbb{G}_t}{d\mathbb{Q}} \right]\nonumber\\
&=& \int {dS_t \,q\left( {S_t ;S_0 ,K,T} \right)} \,\mathbb{
E}^{\mathbb{Q}}\left[\left.\sigma _{t} ^2\right|S_t\right]\nonumber\\ &=& \int {dS_t
\,q\left({S_t ;S_0 ,K,T} \right)}
\,v_\ell(S_t,t)\end{eqnarray*}

where we further define

$$q\left( {S_t,t ;S_0 ,K,T} \right): = \frac{{\,p\left( {S_t,t;S_0 }
\right)\,S_t ^2\, \Gamma _{BS} (S_t ,t)}}{{\mathbb{ E}\left[ {S_t ^2
\,\Gamma _{BS} (S_t ,t)} \right]}}
$$

and
$v_\ell(S_t,t)=\mathbb{ E}^{\mathbb{Q}}\left[\left.\sigma _{t} ^2  \right|S_t\right]$ is
the local variance.

### Brownian Bridge-like density

- <span>$q\left( {S_t,t;S_0 ,K,T} \right)$ looks like a Brownian Bridge
density for the stock price:</span>

    - <span> $p\left(
{S_t,t;S_0 } \right)$ has a delta-function peak at $S_0$ at time $0$
and</span>

    - <span>$\Gamma _{BS} (S_t,t)$ has a delta-function peak at $K$ at
expiration $T$.</span>

For convenience in what follows, we now rewrite equation
[(8)](#eq8:brownianbridgeS) in terms of $x_t := \log {\left( S_t /
S_0 \right )}$:

<a name="eq9:brownianbridgex"></a>(9)
$$v\left(
t \right) = \int {dx_t \,q\left( {x_t ,t;x_T,T} \right)} \,v_\ell(x_t
,t) $$

### The quasi-Brownian Bridge density

```r
p <- function(St,t,S0,sigma){
    x <- log(St/S0)
    sig <- sigma*sqrt(t)
    return(dnorm(x/sig-sig/2)/sig/St)
   }
    
gammaBS <- function(S0,K,T,sigma){
    x <- log(S0/K)
    sig <- sigma*sqrt(T)
    d1 <- x/sig+sig/2
    return(dnorm(d1)/sig/S0)
   }

q <- function(St,t,S0,K,T,sigma){p(St,t,S0,sigma)*gammaBS(St,K,T-t,sigma)*St^2}

q1 <- function(x,t){q(exp(x),t,S0=1,K=1.3,T=1,sigma=.2)}
q1Vec <- function(k,t){sapply(t,function(t){q1(k,t)})} # Vectorized version for plotting
```

### The most-likely-path approximation

- <span>Figure 1 shows what $q\left( {x_t ,t;x_T,T} \right)$ looks
like in the case of a 1 year European option struck at 1.3 with a flat
20% volatility. </span>

    - <span>We see that $q\left( {x_t ,t;x_T,T} \right)$ peaks on a line
(which we will denote by $\tilde x_t$ and call the
<span>*most-likely-path*</span>) joining the stock price today with the
strike price at expiration.</span>

    - <span> Moreover, the density looks roughly symmetric around the peak.
This suggests an expansion around the peak $\tilde x_t$ (at which the
derivative of $q\left( {x_t ,t;x_t,T} \right)$ with respect to $x_t$ is
zero). </span>

This motivates us to write:

<a name="eq10:approx"></a>(10)
$$q\left( {x_t ,t;x_T,T} \right) \approx q( \tilde x_t,t;x_T,T)
 + \frac{1}{2}\left( x_t  - \tilde
x_t \right)^2 \left. {\frac{\partial ^2 q}{\partial x_t ^2 }}
\right|_{x_t  = \tilde x_t} $$

In practice, the local variance $v_\ell(x_t,t)$ is typically not so far
from linear in $x_t$ in the region where $q\left( {x_t
,t;x_T,T} \right)$ is significant so we may further write

<a name="eq11:sigapprox"></a>(11)
$$v_\ell(x_t,t) \approx v_\ell(\tilde x_t,t)
 + \left( x_t  - \tilde
x_t \right) \left. {\frac{\partial v_\ell }{\partial x_t }}
\right|_{x_t  = \tilde x_t} $$

Substituting [(10)](#eq10:approx) and [(11)](#eq11:sigapprox) into the integrand in
equation [(9)](#eq9:brownianbridgex) gives

$$v\left( t \right) \approx v_\ell(\tilde x_t,t)$$ 

and we may rewrite
equation [(6)](#eq6:explicitimpvol) as

<a name="eq12:impvolfromlocvol"></a>(12)
$$\sigma_{BS}(K,T)^2 \approx \frac {1}{T} \int_0^T {v_\ell(\tilde
x_t)\, dt}$$

### The MLP approximation in words

In words, equation [(12)](#eq12:impvolfromlocvol) says that

<blockquote>
<div style="background-color:#E8E8E8; color:#000000; font-style: normal; ">
<br>
The Black-Scholes implied variance of an option with strike $K$ is given
approximately by the integral from valuation date ($t=0$) to the
expiration date ($t=T$) of the local variances along the path
$\tilde x_t$ that maximizes the Brownian Bridge density $q\left(
{x_t ,t;x_T,T} \right)$.

<br>
<br>

</div>

</blockquote>



- <span> Note that in practice, it’s not trivial to compute the most
likely path $\tilde
x_t$.</span>

    -   <span>[Adil Reghai]<sup id="cite_ref-Reghai" class="reference"><a href="#cite_note-Reghai"><span>[</span>7<span>]</span></a></sup> describes an efficient fixed-point algorithm to do
    this.</span>

### A fixed point algorithm for finding the most likely path

For each log-strike $k:=\log(K/S_0)$, we approximate the most likely
path $\tilde x_t$ as $$\tilde x_t = \frac{w_t}{w_T}\,k$$ with
$$w_t=\int_0^t\,ds \,{v_\ell(\tilde x_s)}$$

Obviously, this definition is circular. We solve by starting with the
straight line $$\tilde x_t = \frac{t}{T}\,k$$ as our initial guess and
iterating until the path doesn’t change. This iteration is extremely
fast in practice.

### Problems with the MLP approximation

- The MLP approximation should work well if local variance is
roughly linear in the neighborhood of the peak of $q(\cdot)$.

    - Conversely, if there is substantial curvature, the approximation
does not work well.

        - For example, at-the-money if there is curvature and little or no
skew (as in FX).

    - The MLP approximation is too rough for longer dates when the
density $q(\cdot)$ is not so highly peaked.

   - It’s not clear that the peak of the density is the appropriate most-likely-path. For example, Reghai actually approximates the conditional expectation:

$$\tilde S_t := {\mathbb{E} }\left[S_t|S_T=K\right]$$ 

- <span>[Guyon and Henry-Labordère]<sup id="cite_ref-GuyonHenryLabordere" class="reference"><a href="#cite_note-GuyonHenryLabordere"><span>[</span>4<span>]</span></a></sup> show how to correct for curvature (it is
rather slow and complicated).</span>

### Variational MLP

More recently, [Gatheral and Wang]<sup id="cite_ref-GatheralWang" class="reference"><a href="#cite_note-GatheralWang"><span>[</span>3<span>]</span></a></sup> came up with another version of MLP where the
most-likely-path is now uniquely defined.

Define $$f(x(t),t)=\left.\partial_t \log \sigma(x,t)\right|_{x =x(t)}$$
Then
$$\sigma_{\text{BS}}(k,T) \approx \frac{\frac{1}{T}\,\int_0^T\,\sigma(\tilde x(t),t)\,\exp\left\{\int_0^t\,f(\tilde x(s),s)\,ds\right\}\,dt}
 {\sqrt{\frac{1}{T}\,\int_0^T\exp\left\{2\,\int_0^t\,f(\tilde x(s),s)\,ds\right\}\,dt}}.
$$ 

The most-likely-path $\tilde x(t)$ given (implicitly again) by

<a name="eq13:fixedpointx"></a>(13)
$$\tilde x(t)=k\,\frac{\int_0^t\,du\,\sigma(\tilde x(u),u)\,\exp \left\{\int_0^u\,f(\tilde x(s),s)\,ds\right\}}{\int_0^T\,du\,\sigma(\tilde x(u),u)\,\exp \left\{\int_0^u\,f(\tilde x(s),s)\,ds\right\}}.
$$

### Local variance in the Heston model

Recall the Heston SDE:

<a name="eq14:HestonSDE"></a>(14)
\begin{eqnarray*}
dx_t &=&  - \frac {v_t}{2} dt + \sqrt { v_ t} \, dZ_t \nonumber
\\
dv_t &=&  - \lambda \,(v_t- \bar v) dt + \rho\, \eta\, \sqrt {v_t }\,
dZ_t + \sqrt{1 - \rho ^2}\, \eta\, \sqrt {v_t }\,
dZ_t^\perp\end{eqnarray*}

where $dZ_t$ and $dZ_t^\perp$ are orthogonal. 

Eliminating $\sqrt {v_t}
dZ_t$, we get

<a name="eq15:veqn"></a>(15)
$$dv_{ t}   =  - \lambda \,(v_t - \bar v)\,dt +\rho \,\eta \,\left(
dx_t + \frac{1}{2}\,v_t\, dt \right)\\+ \sqrt{1 - \rho ^2} \,\eta\,
\sqrt {v_t }\,dZ_t^\perp $$

Our strategy will be to compute local variances in the Heston model and
then integrate local variance from valuation date to expiration date to
approximate the BS implied variance using equation
[(12)](#eq12:impvolfromlocvol).

### Expected variance

First, consider the unconditional expectation $\hat v_s$ of the
instantaneous variance at time $s$. Solving equation [(15)](#eq15:veqn) gives

$$\hat v_s \, = \, (v_0-\bar v)\,e^{-\lambda \,s}\,+\,\bar v.$$ 

Then
define the expected total variance to time $t$ through the relation

$$\hat w_t \, := \, \int _0 ^t \hat v_s ds \,=\, (v_0-\bar v)
\left\{\frac {1-e^{-\lambda \,t}}{\lambda} \right\} \,+\,\bar v \,t.$$

Finally, let $u_t := \mathbb{ E}[v_t \left| x_T]\right.= v_\ell(x_T,T) $ be the
expectation of the instantaneous variance at time $t$ conditional on the
final value $x_T$ of $x$.

### An ansatz

According to Wikipedia:

> “... an ansatz is an educated guess that is verified later by its
> results.”

Without loss of generality, assume $x_0=0$. Then, the ansatz is that

$$\mathbb{ E}[ {x_s \left| {x_T } \right.} ] = x_T \frac{{\hat w_s
}}{{\hat w_T }}$$

where $ \hat w_t  := \int_0^t {ds} \,\hat v_s$ is the
expected total variance to time $t$ .

 To see that this ansatz is
plausible, note that

$$\mathbb{ E}\left( {x_s } \right) = \mathbb{ E}(x_T )\frac{{\hat
w_s }}{{\hat w_T }} =  - \frac{{\hat w_T }}{2}\frac{{\hat w_s
}}{{\hat w_T }} = - \frac{{\hat w_s }}{2}.$$

If the process for $x_t$
were a conventional Brownian Bridge process, the ansatz would be true;
in this case, it is only approximately true.

### Approximate computation of local variance

Assuming the ansatz to be correct, we may take the conditional
expectation of [(15)](#eq15:veqn) to get:

<a name="eq16:dexpectedlocvar"></a>(16)
$$du_t \,=\,-\lambda ( u_t - \bar v) dt \,+\,\frac{\rho \eta}{2}u_t
dt\,+\, \rho \eta \frac{x_T}{\hat w_T} d \hat w_t \,\\+\, \sqrt{1 -
\rho ^2} \eta \sqrt {v_t }\,\mathbb{ E}[ {dW_t \left| {x_T }
\right.} ] $$

If the dependence of $dW_t$ on $x_T$ is weak or if $
\sqrt{1-\rho^2}$ is very small, we may drop the last term to get

$$du_t \,\approx \,-\lambda' ( u_t - \bar v') dt \,+\, \rho \eta
\frac{x_T}{\hat w_T} \hat v_t dt$$

with
$ \lambda ' = \lambda  - {\rho \eta}/{2}$, $\bar v' = \bar v
{\lambda }/  {\lambda '}$. The solution to this equation is

<a name="eq17:hestonlocalvar"></a>(17)
$$
 v_\ell(x_T,T) =
u_T \approx \hat v'_{ T} \,+\,\rho \eta \,\frac{{x_T }}{{\hat w_T
}}\,\int_0^T {\hat v_s } \,e^{ - \lambda '(T - s)} ds
$$

with $ { \hat v'}_{ s} := \left( {{ v - \bar v'}} \right)e^{ - \lambda'
\,s} {  + \bar v'}$.

### The accuracy of our Heston local volatility approximation

- <span> Equation [(17)](#eq17:hestonlocalvar) gives us an approximate but
surprisingly accurate formula for local variance within the Heston model
(an extremely accurate approximation when $\rho=\pm 1$) .</span>

    - <span> Local variance is approximately linear in $x = \log \left( 
{F}/{K} \right)$.</span>


- <span>To get [(17)](#eq17:hestonlocalvar), we made two approximations: the ansatz and dropping the
last term in [(16)](#eq16:dexpectedlocvar). </span>

- <span> For reasonable parameters, equation [(17)](#eq17:hestonlocalvar) gives
good intuition for the functional form of local variance.</span>

    - <span>When $\rho=\pm1$, it is almost exact.</span>

    - <span>Equation [(17)](#eq17:hestonlocalvar) is in fact exact to first order in
$\eta$ whether or not the ansatz holds or $\sqrt{1-\rho^2}$ is
small.</span>

    - <span>Accordingly, [(17)](#eq17:hestonlocalvar) is also consistent with the
general perturbative expansions of [Lewis]<sup id="cite_ref-Lewis" class="reference"><a href="#cite_note-Lewis"><span>[</span>6<span>]</span></a></sup>.</span>

###A special case of Heston: The Heston-Nandi model

From equation [(17)](#eq17:hestonlocalvar), if $\rho=\pm1$ and if $v=\bar v$, Heston local variance should be well approximated by 

<a name="eq:HestonNandiLV"></a>(18)
$$
\begin{eqnarray*}
 v_\ell(x_T,T) &=& \left[\hat v'_{ T} \,-\, \eta
 \,\frac{{x_T }}{{w_T }}\,\int_0^T {\hat v_s } \,e^{
 - \lambda '(T - s)} ds \right]^+\nonumber\\
 &=&\left[(v-\bar v') e^{ - \lambda' T} +\bar v' \,-\,\eta \, x_T
 \,\left\{ \frac{1-e^{-\lambda' T}}{\lambda' T } \right\}\right]^+
\end{eqnarray*}
$$

with $ \lambda ' = \lambda  +\frac { \eta}{2}$, $\bar v' = \bar v
\frac {\lambda }  {\lambda '}$.  The whole expression must be
bounded below by zero - all stock prices above the critical stock
price at which the local variance reaches zero are unattainable.

$$
\begin{eqnarray*}
v&=& 0.04\\ \bar v &=& 0.04\\ \lambda&=&10\\ \eta&=&1\\\rho&=&-1.
\end{eqnarray*}
$$

### Implied volatility in the Heston model

- Having computed approximate local variance in the Heston model, we now apply the MLP appproximation to get implied variance.

    - <span> We need to integrate the Heston local variance along the most
probable stock price path joining the initial stock price to the strike
price at expiration (the one which maximizes the Brownian Bridge
probability density).</span> 

Then, the Black-Scholes implied variance is
given by

<a name="eq:impliedvariance"></a>(19)
$$\sigma_{BS}(K,T)^2 \approx \frac {1}{T} \int_0^T v_\ell({\tilde
x}_t,t)\, dt
$$

where $\left \{ \tilde x_t \right \}$ is the most likely path (as
defined earlier).

Recall that the Brownian Bridge density $q\left( {x_t ,t;x_T,T}
\right)$ is roughly symmetric and peaked around $\tilde x_t$, so $
\mathbb{ E}\left[ {x_t - \tilde x_t\left| {x_T} \right.} \right]
\approx 0$. Applying the ansatz once again, we obtain

$$\tilde x_t =\mathbb{ E}\left[ { \tilde x_t\left| {x_T} \right.}
\right] =\mathbb{ E}\left[ { \tilde x_t-x_t\left| {x_T} \right.}
\right]\,+\,\mathbb{ E}\left[ { x_t\left| {x_T} \right.} \right]
\approx \frac{\hat w_t}{\hat w_T} x_T$$

We substitute this expression
back into equations [(17)](#eq17:hestonlocalvar) and [(19)](#eq:impliedvariance) to
get

<a name="eq:hestonimpliedvariance"></a>(20)
\begin{eqnarray*}
\sigma_{BS}(K,T)^2 &\approx& \frac {1}{T}\int _0 ^T u_t(\tilde
x_t) dt \nonumber
\\
&\approx&\,\frac {1}{T}\int _0 ^T {\hat v'_{ t}dt \,+\,\rho \eta
\,\frac{{x_T }}{{\hat w_T }}\frac {1}{T}\int _0 ^T dt  \,\int_0^t
{\hat v_s } \,e^{ - \lambda '(t - s)} ds}\nonumber\\
\end{eqnarray*}

### The term structure of implied volatility in the Heston model

The at-the-money term structure of BS implied variance in the Heston
model is obtained by setting $x_T=0$ in equation
[(20)](#eq:hestonimpliedvariance). Performing the integration explicitly
gives

<!--<a name="eq20:approxatmvar"></a>(20)-->
\begin{eqnarray*}
\left.{\sigma_{BS}(K,T)^2}\right|_{K=F_T} &\approx&\frac {1}{T}\int
_0 ^T {\hat v'_{ t}dt} \\&=& \frac {1}{T}{\int_0^T \left[\left( {{
v - \bar v'}}\right) e^{ - \lambda' \,t}  + \bar v' \right] dt}
\nonumber \\&=&\left( {{ v - \bar v'}}\right)\frac{1- e^{ -
\lambda' \,T}}{\lambda' \,T} + \bar v'.
\end{eqnarray*}

We see that in the Heston model, the at-the-money Black-Scholes implied
variance 

$$\left.{\sigma_{BS}(K,T)^2}\right|_{K=F_T}\,\rightarrow \,v$$


as $T \rightarrow 0$ and as $T \rightarrow \infty$, the at-the-money
Black-Scholes implied variance reverts to $\bar v'$.

### The implied volatility skew in the Heston model

- <span>By inspection of [(19)](#eq19:hestonimpliedvariance) the implied variance
skew in the Heston model is approximately linear in the correlation
$\rho$ and the volatility of volatility $\eta$.</span>

- <span>When $ v_0 = \bar v$, $ \hat v_s  = \bar v$ and
$\hat w_t  = \bar v\,t$. </span>

    - <span>The most likely path $\tilde x_t
\approx \frac{t}{T} x_T$ is exactly a straight line in log-space between
the initial stock price on valuation date and the strike price at
expiration. </span>

Performing the integrations in equation [(20)](#eq:hestonimpliedvariance)
explicitly, we get

<a name="eq21:hestonsmile"></a>(21)
\begin{eqnarray*}
\sigma_{BS}(K,T)^2 &\approx&  \frac{\hat w_T ^\prime}{T} \,+\,\rho
\eta \frac{x_T}{T^2} \int _0 ^T \, dt\,\frac {1}{T}\int_0^t {\,e^{
- \lambda '(t - s)} } ds \nonumber
\\ &=& \frac{\hat w_T ^\prime}{T}  +\,\rho \eta\,  \frac{x_T}{{\lambda ' T}}\left\{ {1 -
\frac{{\left( {1 - e^{ - \lambda 'T} } \right)}}{{\lambda 'T}}}
\right\} \end{eqnarray*}

### Qualitative features of the ATM skew

- <span> From [(21)](#eq21:hestonsmile), we see that the implied variance skew
${\frac{\partial }{\partial x_t}}
{\sigma_{BS}(K,T)^2}$ is *independent* of the level of instantaneous
variance $v$ or long-term mean variance $\bar v$.</span>

    - <span>This remains approximately true even when $ v \neq \bar v$.</span>

- <span>It follows that we now have a fast way of calibrating the Heston
model to observed implied volatility skews. </span>

    - <span>Just two expirations would in principle allow us to determine
$\lambda'$ and the product $\rho
\eta$. </span>

    - <span> We can then fit the term structure of volatility to determine the
long term mean variance $\bar v$ and the instantaneous variance
$v_0$.</span>


- <span> The curvature of the skew (not discussed here) would allow us to
determine $\rho$ and $\eta$ separately.</span>


- <span> We note that as we increase either the correlation $\rho$ or the
volatility of volatility $\eta$, the skew increases.</span>

### Asymptotic analysis

- <span>The very short-dated skew is independent of $\lambda$ and $T$:

$$
\begin{eqnarray*}
\frac{\partial } {\partial x_t} \sigma_{BS}(K,T)^2 &=&\rho
\eta\, \frac{1}{{\lambda ' T}}\left\{ {1 - \frac{{\left( {1 - e^{ -
\lambda 'T} } \right)}}{{\lambda 'T}}}
\right\}\\&\rightarrow&\,\frac{\rho \eta}{2}\textrm{ as } T
\rightarrow 0.
\end{eqnarray*}$$ </span>

- <span> The long-dated skew is inversely proportional to $T$:

$$
\begin{eqnarray*}
\frac{\partial } {\partial x_t} \sigma_{BS}(K,T)^2 &=&\rho
\eta\,
 \frac{1}{{\lambda ' T}}\left\{ {1 - \frac{{\left( {1 - e^{ -
\lambda 'T} } \right)}}{{\lambda 'T}}} \right\}\\
&\sim&\frac{\rho
\eta}{\lambda ' T}\textrm{ as } T \rightarrow \infty.
\end{eqnarray*}
$$ </span>

- <span> Finally, increasing $\eta$ causes the curvature of the implied
volatility skew (related to the kurtosis of the risk-neutral density) to
increase but we haven’t shown that here.</span>

### Dynamics of the volatility skew under local volatility

Recall formula (1.10) from TVS for local volatility in terms of implied
volatility:

$$v_\ell  = \frac{{\frac{{\partial { {  w}} }}{{\partial T}}}}{{1 -
\frac{k}{{{ {  w}} }}\frac{{\partial { { w}} }}{{\partial k}} +
\frac{1}{4}\left( { - \frac{1}{4} - \frac{1}{{{ {  w}} }} +
\frac{{k^2 }}{{{ {  w}} ^2 }}} \right)\left( {\frac{{\partial { {
w}} }}{{\partial k}}} \right)^2  + \frac{1}{2}\frac{{\partial ^2 {
{  w}} }}{{\partial k^2 }}}}.$$ 

Differentiating with respect to $k$ and
considering only the leading term in $\frac{\partial  w }{\partial k}$
(which is small for large $T$) , we find

$$\frac{\partial  v_\ell }{\partial k}  \approx \frac{\partial
}{\partial T}\frac{\partial w }{\partial k}\,+\,
\frac{1}{w}\frac{\partial w }{\partial T}\frac{\partial w
}{\partial k}.$$ 

That is, the local variance skew
$\frac{\partial  v_{loc}
}{\partial k}$ decays with the BS implied total variance skew
$\frac{\partial w }{\partial k}$.

### Dynamics of the volatility skew under local volatility

- <span> To get the forward volatility surface in a local volatility
model, we integrate over the local volatilities from the (forward)
valuation date to the expiration of the option along the most likely
path.</span>

    - <span>The forward volatility surface will be substantially flatter than
today’s because the forward local volatility skews are all
flatter.</span>

- <span>Contrast this with a stochastic volatility model where implied
volatility skews are approximately time-homogeneous.</span>

    - <span>Local volatility models imply that future BS implied volatility
surfaces will be flat (relative to today’s).</span>

    - <span>Stochastic volatility models imply that future BS implied
volatility surfaces will look like today’s.</span>

### The SVI parameterization

In the SVI (“Stochastic Volatility Inspired”) parameterization, for each
expiration, we write

<a name="eq22:SVI"></a>(22)
$$\sigma^2_{BS}(k)=a+b\,\left\{\rho\,(k-m)+\sqrt{(k-m)^2+\sigma^2}\right\}
$$

where the coefficients $a$, $b$, $\rho$, $\sigma$ and $m$ depend on the
expiration.

- <span>SVI is just a <span>*parameterization*</span>, there are no underlying
dynamics associated with it. </span>

- <span>In [Gatheral and Jacquier]<sup id="cite_ref-GatheralJacquier" class="reference"><a href="#cite_note-GatheralJacquier"><span>[</span>2<span>]</span></a></sup>, we show that it is possible to guarantee no static
arbitrage when fitting SVI.</span>

    - <span>The functional form [(22)](#eq22:SVI) is fitted to all expirations
simultaneously subject to the constraint that there should be no
calendar spread or butterfly arbitrage. </span>


- <span>Total implied variance may then be interpolated between slices to
give a smooth surface.</span>

    - <span> Here we use Stineman monotonic spline interpolation. </span>

Code to generate Table 1.

```r
sviVarianceATM <- function(sviParams){

    a <- sviParams$a
    b <- sviParams$b
    rho <- sviParams$rho
    m <- sviParams$m
    sig <- sviParams$sig
    
    var <-a + b*(-m*rho+sqrt(m^2+sig^2))
    return(var)
    
    }

sviSkewATM <- function(sviParams){

    b <- sviParams$b
    rho <- sviParams$rho
    m <- sviParams$m
    sig <- sviParams$sig
    
    skew <- b*(rho-m/sqrt(m^2+sig^2))
    return(skew)
    
    }

nexp <- length(texp)
atmVars <- numeric(nexp)
atmSkews <- numeric(nexp)

for (i in 1:nexp){
    
    atmVars[i] <- sviVarianceATM(sviMatrix[i,])/texp[i]
    atmSkews[i] <- sviSkewATM(sviMatrix[i,])/texp[i]
    
    }

atmVols <- sqrt(atmVars)
```

### Term structure of ATM skew

A function to draw the approximate term structure of Heston skew:

```r
skewHeston <- function(skewParams,texp){

    rhoeta <- skewParams[1]
    lam <- skewParams[2]
    lt <- lam*texp
    skew <- rhoeta/lt*(1-(1-exp(-lt))/lt)	
    return(skew)
    
    }
```

Now fit this skew function to the empirical points:

```r
skewObjective1 <- function(skewParams){
    
    skewModel <- skewHeston(skewParams,texp[-1])
    skewEmpirical <- atmSkews[-1]
    obj <-sum((skewModel-skewEmpirical)^2)
    return(obj*1000000)
    
    }

res1 <- optim(c(-.7*.39,1.15),skewObjective1)
fit1 <- res1$par 
```

Fit again dropping the first three points:

```r
skewObjective2 <- function(skewParams){
    
    skewModel <- skewHeston(skewParams,texp[-(1:3)])
    skewEmpirical <- atmSkews[-(1:3)]
    obj <-sum((skewModel-skewEmpirical)^2)
    return(obj*1000000)
    
    }

res2 <- optim(c(-.7*.39,1.15),skewObjective2)
fit2 <- res2$par
```

Now fit a power law:

```r
skewObjectivePL <- function(alpha){
    
    skewModel <- atmSkews[5]*(texp[5]/texp[-1])^alpha
    skewEmpirical <- atmSkews[-1]
    obj <-sum((skewModel-skewEmpirical)^2)
    return(obj*1000000)
    
    }

resPL <- optim(.4,skewObjectivePL)
fitPL <- resPL$par
```

### Term structure of ATM variance

A function to approximate Heston ATM variance term structure:

```r
varHeston <- function(varParams,texp){
    
    v <- varParams[1]
    vbar <- varParams[2]
    lam <- varParams[3]
    lt <- lam*texp
    var <- vbar + (v-vbar)*(1-exp(-lt))/lt	
    return(var)
    
    }
```

Fit empirically ATM variances dropping the first point:

```r
varObjective <- function(varParams){
    
    varModel <- varHeston(varParams,texp[-1])
    varEmpirical <- atmVars[-1]
    obj <-sum((varModel-varEmpirical)^2)
    return(obj*1000000)
    
    }

(res <- optim(c(0.04,0.04,1.15),varObjective))
fit <- res$par
```

### Comparison of Heston fit to empirical surface: 3D plot

First a function to return Heston implied vols for a vector of log-strikes $k$ and expirations $t$.

```r
paramsHeston050915 <- list(
    v= 0.0174,
    vbar=0.0354,
    eta=0.3877,
    rho=-0.7165,
    lambda=1.3253)

k <- seq(-.5,.5,0.01) # Vector of log-strikes
t <- seq(0.04,1.74,0.02) # Vector of times

vol1 <- function(k,t){bsvol(phiHeston(paramsHeston050915),k,t)}
vol2 <- function(k,t){sapply(k,function(k){vol1(k,t)})}
vol <- function(k,t){sapply(t,function(t){vol2(k,t)})}
```

Now built the Heston volatility surface:

```r
zvol <- vol(k,t)

# First the analytical formula: 
z <- (zvol>.05)*zvol+(zvol <= 0.05)*.05

# Add colors
nbcol <- 100
color <- rainbow(nbcol,start=.0,end=.2)
nrz <- nrow(z)
ncz <- ncol(z)
# Compute the z-value at the facet centres
zfacet <- z[-1, -1] + z[-1, -ncz] + z[-nrz, -1] + z[-nrz, -ncz]
# Recode facet z-values into color indices
facetcol <- cut(zfacet, nbcol)
```

Plot the Heston surface:

```r
persp(k, t, z, col=color[facetcol], phi=30, theta=30, 
    r=1/sqrt(3)*20,d=5,expand=.5,ltheta=-135,lphi=20,ticktype="detailed",
    shade=.5,border=NA,xlab="Log-strike k",ylab="Expiration t",zlab="Implied volatility",zlim=c(.05,.5))
```

Next build the empirical surface:

```r
z2 <- t(volTVS(k,t)) # Array of empirical volatilities

# Add colors
nbcol <- 100
color2 <- rainbow(nbcol,start=.4,end=.6)
nrz <- nrow(z)
ncz <- ncol(z)
# Compute the z-value at the facet centres
zfacet <- z[-1, -1] + z[-1, -ncz] + z[-nrz, -1] + z[-nrz, -ncz]
# Recode facet z-values into color indices
facetcol <- cut(zfacet, nbcol)
```

and plot it:

```r
persp(k, t, z2, col=color2[facetcol], phi=30, theta=30, 
    r=1/sqrt(3)*20,d=5,expand=.5,ltheta=-135,lphi=20,ticktype="detailed",
    shade=.5,border=NA,xlab="Log-strike k",ylab="Expiration t",zlab="Implied volatility",zlim=c(.05,.5))
```

Now plot them together with the empirical surface shifted up by two vol points (to make the comparison easier):

```r
persp(k, t, z, col=color[facetcol], phi=30, theta=30, 
    r=1/sqrt(3)*20,d=5,expand=.5,ltheta=-135,lphi=20,ticktype="detailed",
    shade=.5,border=NA,xlab="Log-strike k",ylab="Expiration t",zlab="Implied volatility",zlim=c(.05,.5))
par(new=T)
persp(k, t, z2+.02, col=color2[facetcol], phi=30, theta=30, 
    r=1/sqrt(3)*20,d=5,expand=.5,ltheta=-135,lphi=20,ticktype="detailed",
    shade=.5,border=NA,xlab="Log-strike k",ylab="Expiration t",zlab="Implied volatility",zlim=c(.05,.5))
```

### Summary of Lecture 5

- We showed how to approximate local volatility in the Heston model.
   - The approximation is accurate when $\rho = \pm 1$ (The Heston-Nandi model in particular).  Thus we have a local volatility model that generates almost the same European option prices as the Heston model in this case.
   
   
- The Black-Scholes implied variance of an option with strike $K$ is given
approximately by the integral from valuation date ($t=0$) to the
expiration date ($t=T$) of the local variances along the path
$\tilde x_t$ that maximizes a Brownian Bridge-like density.

   

-  We used this most-likely-path approximation to approximate implied volatility in the Heston model.
    - We were then able to approximate the term structures of ATM variance and ATM skew.
    
    
-  We were then able to deduce that the Heston model generates a volatility surface that does not look like the empirically observed volatility surface.
